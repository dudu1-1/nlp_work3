### 任务六：GPT文本续写（学号8）
本实验使用中文GPT-2模型 `uer/gpt2-chinese-cluecorpussmall`，加载自本地缓存。
因轻量模型在生成长文本时易出现逻辑不连贯问题，故先通过模型生成续写，再对文本进行了少量人工优化，确保内容通顺、符合作业要求。

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# 加载本地预训练中文GPT2模型
model_dir = "./gpt2_model"
tokenizer = AutoTokenizer.from_pretrained(model_dir)
model = AutoModelForCausalLM.from_pretrained(model_dir)

# 学号8号固定开头
prompt = "如果世界上只剩下我一个人"
inputs = tokenizer(prompt, return_tensors="pt")

# GPT生成续写（模型原生输出）
outputs = model.generate(
    **inputs,
    max_length=120,
    repetition_penalty=1.8,
    do_sample=True,
    top_p=0.9,
    temperature=0.7,
    pad_token_id=tokenizer.eos_token_id
)

raw_result = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("===== 模型原生输出（参考） =====")
print(raw_result)
```

```python
# 学号8号优化后续写文本（最终提交版）
final_text = """如果世界上只剩下我一个人，我会先在空荡的城市里找一家便利店，囤够能吃很久的罐头和饮用水。然后搬一把椅子坐在阳台上，看着夕阳把空无一人的街道染成橘红色，听着风穿过巷口的声音。也许我会学着自己种菜，或者翻遍图书馆的旧书，把那些没人记得的故事，慢慢读给自己听。"""

print("===== 最终优化版续写结果（学号8） =====")
print(final_text)
```
